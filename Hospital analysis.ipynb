```python
%sql
-- Create a temporary view 'hospital_data' combining appointment, patient, doctor, treatment, and billing information
create or replace temp view hospital_data as
select 
  a.appointment_date, -- Appointment date
  a.appointment_id, -- Unique appointment identifier
  p.patient_id, -- Unique patient identifier
  concat(p.first_name,' ',p.last_name) as patient_name, -- Patient's full name
  p.gender, -- Patient gender
  p.date_of_birth, -- Patient date of birth
  d.doctor_id, -- Unique doctor identifier
  concat(d.first_name,' ',d.last_name) as doctor_name, -- Doctor's full name
  d.specialization, -- Doctor's specialization
  a.reason_for_visit as treatment_name, -- Reason for visit (used as treatment name)
  a.status, -- Appointment status
  b.bill_date, -- Billing date
  b.amount -- Billing amount
from
  hospital_management_datasets.default.appointments as a
join 
  hospital_management_datasets.default.patients p on a.patient_id = p.patient_id -- Link appointments to patients
join
  hospital_management_datasets.default.doctors d on a.doctor_id = d.doctor_id -- Link appointments to doctors
join
  hospital_management_datasets.default.treatments t on a.appointment_id = t.appointment_id -- Link appointments to treatments
join
  hospital_management_datasets.default.billing b on a.patient_id = b.patient_id -- Link patients to billing records
```

```python
%sql
-- Retrieve all columns from the hospital_data view, which combines appointment, patient, doctor, treatment, and billing information
select 
  *
from hospital_data
```

```python
-- Calculate the total revenue generated from all billing records in the hospital_data view
select 
  sum(amount) as total_revenue
from hospital_data
```

```python
%sql
-- Calculate total revenue generated by each doctor's specialization
select 
  specialization, -- Doctor's specialization
  sum(amount) as total_revenue -- Total billing amount for the specialization
from
  hospital_data
group by specialization  
order by total_revenue -- Sort by total revenue in ascending order
```

```python
-- Count the total number of unique appointments in the hospital_data view
select count(distinct appointment_id) as total_appointments
from hospital_data
```

```python
%sql
-- Count the number of times each treatment was performed
select 
  treatment_name, 
  count(*) as total_count
from hospital_data
group by treatment_name
order by total_count
```

```python
%sql
-- Calculate the average billing amount across all records
select avg(amount) as average_bill
from hospital_data
```

```python
%sql
-- Count distinct appointments per doctor specialization
select 
  specialization,
  count(distinct appointment_id) as total_appointment_per_specialization
from hospital_data
group by specialization
order by total_appointment_per_specialization desc
```

```python
-- Top 5 doctors by total revenue generated from billing amounts
SELECT doctor_name,
       ROUND(SUM(amount),2) AS revenue
FROM hospital_data
GROUP BY doctor_name
ORDER BY revenue DESC
LIMIT 5
```

```python
-- Top 10 doctors by number of distinct appointments
select 
  doctor_name,
  count(distinct appointment_id) as number_of_appointment
from hospital_data
group by doctor_name
order by number_of_appointment desc
limit 10
```

```python
-- Sum billing amounts by treatment name to identify top revenue-generating treatments
select  
  treatment_name,
  sum(amount) as total_Amount
from hospital_data
group by treatment_name
order by total_Amount desc
```

```python
-- Calculate monthly revenue by summing billing amounts, rounded to 2 decimals
select 
  month(bill_date) as month,
  round(sum(amount), 2) as revenue
from hospital_data
group by month
order by month
```

```python
-- Count distinct appointments per date to analyze daily appointment volume
select appointment_date,
       count(distinct appointment_id) as total_appointment
from hospital_data
group by appointment_date
order by total_appointment desc
```

```python
%sql
-- Count distinct patients by gender
select 
  gender,
  count(distinct concat(first_name, ' ', last_name)) as patient_count
from hospital_management_datasets.default.patients
group by gender
```
